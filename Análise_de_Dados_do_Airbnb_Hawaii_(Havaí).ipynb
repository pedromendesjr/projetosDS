### Limpando outliers

Outliers são os valores atípicos presentes em datasets que podem prejudicar a análise dos dados. Existem formas diferentes de identificar a presença dos outliers, sendo as mais comuns o IQR score, z-score, gráficos de boxplots, teste de Grubb e entre outros. 

Para o nosso caso, iremos utilizar o conceito do IQR score nas variáveis `price`, `minimum_nights`, `number_of_reviews` e `reviews_per_month`. 

O IQR score (interquartile range) consiste em: 
* Defini-se o IQR como sendo a diferença entre o valor do 3o quartil (Q3) pelo 1o quartil (Q1). 
* Calcula-se o limite inferior como sendo: Q1 - 1.5 x IQR
* Calcula-se o limite superior como sendo: Q3 + 1.5 x IQR
* Os valores que estiverem fora dos limites inferiores e superiores serão considerados outliers. 

```python
# Definindo o IQR e os limites inferiores e superiores da variável "price"
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
iqr_price = q3_price - q1_price

inf_price = q1_price - (1.5*iqr_price)
sup_price = q3_price + (1.5*iqr_price)

# Resultados:
print('IQR Price: {}'.format(iqr_price))
print('Limite Inferior: {}'.format(inf_price))
print('Limite Superior: {}'.format(sup_price))
```

Sendo assim, teremos um limite máximo de 770,5 no `price` e, mesmo que o limite inferior tenha resultado num número negativo, ainda iremos tratar os dados retirandos os valores "0" de algumas propriedades.

```python
# Definindo o IQR e os limites inferiores e superiores da variável "minimum_nights"
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
iqr_minimum_nights = q3_minimum_nights - q1_minimum_nights

inf_minimum_nights = q1_minimum_nights - (1.5*iqr_minimum_nights)
sup_minimum_nights = q3_minimum_nights + (1.5*iqr_minimum_nights)

# Resultados:
print('IQR Minimum Nights: {}'.format(iqr_minimum_nights))
print('Limite Inferior: {}'.format(inf_minimum_nights))
print('Limite Superior: {}'.format(sup_minimum_nights))
```

```python
# Bloxplot antes da retirada de outliers:
fig, (ax1, ax2) = plt.subplots(2,1,figsize = (12,5))
df.price.plot(kind='box', ax = ax1 , vert=False)
ax1.set_title('Dataset Original - Variável Price')
df.minimum_nights.plot(kind='box',ax = ax2, vert=False)
ax2.set_title('Dataset Original - Variável Minimum_Nights')

plt.tight_layout()
```

### Retirada dos outliers em 'price' e 'minimum_nights'. 

* Na variável `price`, iremos retirar os valores 0 e os valores acima de 770,5.
* Na variável `minimum_nights` iremos retirar os valores acima de 11. 

```python
# PRICE 

outliers_price = len(df[df.price > 770.5]) + len(df[df.price == 0])
print('Outliers de Price: {}'.format(outliers_price))
print('% do dataset perdida em outliers: {} %'.format(round(outliers_price/df.shape[0],3)*100))
```

```python
# Retirando os outliers de price:
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > sup_price].index, axis=0, inplace = True)
```

```python
# MINIMUM_NIGHTS 

outliers_minimum_nights = len(df_clean[df_clean.minimum_nights > sup_minimum_nights])
print('Outliers de Minimum_Nights: {}'.format(outliers_minimum_nights))
print('% do dataset perdida em outliers: {} %'.format(round(outliers_minimum_nights/df_clean.shape[0],2)*100))

# Retirando os outliers de minimum_nights:

df_clean.drop(df_clean[df_clean.minimum_nights > sup_minimum_nights].index, axis=0, inplace = True)
```

```python
# Dataset Original x Dataset Limpo

print('Tamanho dataset original: {}'.format(df.shape))
print('Tamanho dataset atual: {}'.format(df_clean.shape))
```

```python
# Bloxplot após a retirada de outliers:
fig, (ax1, ax2) = plt.subplots(2,1,figsize = (12,5))
df_clean.price.plot(kind='box', ax = ax1 , vert=False)
ax1.set_title('Dataset Atual - Variável Price')
df_clean.minimum_nights.plot(kind='box',ax = ax2, vert=False)
ax2.set_title('Dataset Atual - Variável Minimum_Nights')

plt.tight_layout()
```

Podemos observar que houve uma evolução significativa nos boxplots das variáveis `price` e `minimum_nights`. Por mais que o boxplot ainda apresende outliers, isso deve a questões mais técnicas que não aprofundaremos nesse momento. 

Vamos dar uma olhada nos histogramas agora.

```python
# Histograma após a retirada de outliers:
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (10,5))
df_clean.price.plot(kind='hist', ax = ax1)
ax1.set_title('Dataset Atual - Variável Price')
df_clean.minimum_nights.plot(kind='hist',ax = ax2)
ax2.set_title('Dataset Atual - Variável Minimum_Nights')

plt.tight_layout()
```

Os histogramas também apresentaram uma melhora significativa. Podemos observar melhor essa distribuição dos dados, diferente dos histogramas anteriores onde praticamente todos os dados se concentravam na primeira barra.

## Q6. Qual a correlação existente entre as variáveis?

A correlação é um valor estatística que busca demonstrar o grau de associação entre duas variáveis. Quanto mais próximo de 1 for a correção, maior a associação. Ou seja, a tendência é que essas variáveis "caminhem juntas". Neste tópico, o nosso objetivo é criar uma matriz de correlação entre algumas variáveis e também plotar um heatmap dessa correlação. 

```python
# Criar uma matriz de correção

corr_df = df_clean[['price', 'minimum_nights','reviews_per_month', 'availability_365']].corr(method = 'pearson')
print("A Matriz de Correlação é:")
display(corr_df)

# Mostrar a matriz de correlação

```

```python
# Plotar um heatmap a partir das correlações

plt.figure(figsize=(8,6))
sns.heatmap(corr_df, annot=True)
plt.show()

```

Esse heatmap nos mostra uma baixíssima correlação entre algumas variáveis.
